# Black Hole Merger Tutorial

This tutorial aims to provide usage knowledge for animating a Binary Black Hole Merger using the **cosmicBallet** package. The binary system visualization utilizes surrogate modelling techniques from the surfinBH module and then extrapolates the trajectories of the Black Holes till merger. The gravitational waves can also be plotted for different modes. For a full binary system merger, it is highly recommended to use a full scale numerical relativity solver (such as Einstein Toolkit) for a detailed simulation (as tidal effects, plasma physics etc., cannot be captured with a surrogate technique).

## Import all Necessary Modules from the package

```python
from cosmicBallet.CelestialObjects import BlackHole
import cosmicBallet.utils.Constants as const
from cosmicBallet.MassiveObjectSimulator import BinaryMerger
```

## Define the Black Holes

Let us assume that the Black Holes have masses of $1.45 * 10^6 M_S$ and $10^6 M_S$ where $M_S$ is $1$ solar mass. Let the angular momentum for both black holes be $0$ in all three directions. 

```python
mass = const.MS
bh1 = BlackHole(name = "Black Hole 1",
                mass = 1.45 * mass * 1e6,
                init_position= [0,0,0],
                init_velocity= [0,0,0])
bh2 = BlackHole(name = "Black Hole 2",
                mass = mass * 1e6,
                init_position= [0,0,0],
                init_velocity= [0,0,0])
```

## Initialize and Simulate

Initialize the BinaryMerger object and run a simulation. When initializing the object, ensure that the larger black hole (or neutron star) is the first item in the list as the surrogate model requires that the mass ratio ($q = \frac{m_1}{m_2} \geq 1$, where $m_1$ is the mass of the first celestial object in the list and $m_2$ is the mass of the second object).

```python
merger = BinaryMerger(binary_system=[bh1, bh2])
merger.simulate()
```

## Plot Gravitational Waves

The following modes are available for plotting gravitational waves:

(2,-2), (2,-1), (2,0), (2,1), (2,2)
(3,-3), (3,-2), (3,-1), (3,0), (3,1), (3,2), (3,3)
(4,-4), (4,-3), (4,-2), (4,-1), (4,0), (4,1), (4,2), (4,3), (4,4)

The notation $(l,m)$ is used to label these modes, where $l$ and $m$ are integers. The number $l$ is the angular momentum quantum number, and $m$ is the magnetic quantum number, which specifies the number of nodes in the angular direction and the orientation of the mode, respectively.

- $l$: This number represents the total angular momentum of the mode. It determines the shape of the wavefront and how it spreads out. Higher $l$ values correspond to modes with more complex angular structures.
- $m$: This number is related to the projection of the angular momentum along the axis of symmetry. It specifies how the mode is oriented relative to this axis.

```python
merger.gw_plot(mode=(4,4))
```

## Animate the Binary Black Hole Merger

```python
merger.animate()
```
