# Stellar Precession Tutorial

This tutorial is about the precessing orbits of stars around a Black Hole. The Black Hole is assumed to be unaffected by stars (as it is more dense and more massive) and is stationary. The mass ratio of the Black Hole and the most massive star needs to be at least $10^3$ and the stars have very high velocity for the simulation to work accurately (regime where the assumptions hold).

## Import Necessary Libraries

```python
from cosmicBallet.CelestialObjects import BlackHole, Stars
import cosmicBallet.utils.Constants as const
from cosmicBallet.MassiveObjectSimulator import SchwarzschildSimulator
```

```python
mass = 1.989e30
radius = 6.96e8
speed = 2.98e7
distance = 1.496e10

black_hole = BlackHole(name="Black Hole", mass=mass*1e6,
                       init_position=[0,0,0], init_velocity=[0,0,0])
star1 = Stars(name="A", mass=mass, temperature=3700, init_position=[distance,0,0],
              init_velocity=[0,2*speed,0], radius=radius)
star2 = Stars(name="B", mass=mass, temperature=5700, init_position=[1.5*distance,0.5*distance,0],
              init_velocity=[speed,speed,0], radius=radius)
star3 = Stars(name="C", mass=mass, temperature=7700, init_position=[2*distance,0.5*distance,0],
              init_velocity=[speed,0,speed], radius=radius)
star4 = Stars(name="D", mass=mass, temperature=19700, init_position=[0,-distance,0],
              init_velocity=[-speed,0,0], radius=radius)
```

## Initialize and Simulate

Initialize the simulation object and solve the problem. The problem is solved by default using the 4th Order Forest-Ruth Integration Scheme along with 1st PN Correction Term (to account for high speeds and high (but not extreme) gravitational field)

```python
simulator = SchwarzschildSimulator(dense_body=black_hole, 
                                   stars=[star1, star2, star3, star4],
                                   time_step=1, simulation_time=1e3)
simulator.solve()
```

## Visualize

Similar to the N-Body Simulations, the visualization is integrated to the simulator class and can be called directly.

### Static Visual

```python
simulator.visualize()
```

### Matplotlib Animation

```python
simulator.visualize(figure_name="Stellar_Precession_MPL", animate=True)
```

### 3D Animation

For simulations where the objects are large enough and the distances short enough, a 3D animation can be used to visualize the orbits

```python
simulator.visualize(visualization_type="animation", figure_name="Stellar_Precession_3D",
                    animate=True, save_figure=True)
```
